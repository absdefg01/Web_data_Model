```python
# verify if there is a pair of root node
def verifyRootNode():
    tab_root = []
    has_r = False
    root = tab_word[0]
    nb = root[0]
    char = root[1]
    nb_last = tab_word[len(tab_word)-1][0]
    char_last = tab_word[len(tab_word)-1][1]
    
    #the nb of the first element is 0 and the nb of the last element = 1
    #and the string of the first element = the string of the last element
    if nb == '0'and nb_last == '1' and char == char_last:
        for i in range(len(tab_word)):
            if tab_word[i][1] == char:
                # 0 : we insert 0 a into tab_root
                if i == 0:
                    if tab_word[i][0] == '0':
                        tab_root.append([tab_word[i],"root"])
                    else:
                        has_r = False
                        break
                else:
                    if tab_word[i][0] == '0':
                        tab_root.append([tab_word[i],"leaf"])
                
                #1 : we remove 0 correspondant
                if i < (len(tab_word)-1):
                    if tab_word[i][0] == '1':
                        if tab_root[-1][1] == "leaf":
                            tab_root.remove(tab_root[-1])
                        else:
                            has_r = False
                            break
                            
                if i == len(tab_word)-1:
                    if tab_word[i][0] == '1':
                        if tab_root[-1][1] == 'root':
                            has_r = True
                        else:
                            has_r = False
                            break
                    else:
                        has_r = False
                        break
    return has_r
```

```python
# verify if the length of 0 and 1 are different : not well formed 
def verifyNbZeroOne():
    # 0
    for j in range(len(tab_word)):
        if (tab_word[j][0] == '0'):
            tab_count_zero.append(tab_word[j][1])
    # 1
    for j in range(len(tab_word)):
        if (tab_word[j][0] == '1'):
            tab_count_one.append(tab_word[j][1])
                           
    if len(tab_count_one) != len(tab_count_zero):
        return False
    else:
        return True
```
